```python
import os

# --- THIẾT LẬP (Giữ nguyên từ lần chạy trước) ---
GITHUB_OWNER_USERNAME = "dhuy198"
GITHUB_REPO_NAME = "SE2025-14.3"
YOUR_GITHUB_USERNAME = "hongk4"
YOUR_GITHUB_EMAIL = "nguyenxuanhong2604@gmail.com" # !! NHỚ THAY EMAIL CỦA BẠN !!

# Đường dẫn đến thư mục chứa ảnh CẦN ĐỔI TÊN bên trong repo
TARGET_IMAGE_DIR = "01_crawling/raw_images/google/xu_li_vi_pham"
# -------------------------------------------------------------------

# 1. LẤY LẠI MÃ TOKEN VÀ CLONE LẠI REPO
from google.colab import userdata
GITHUB_PAT = userdata.get('GITHUB_PAT')
!git config --global user.name "{YOUR_GITHUB_USERNAME}"
!git config --global user.email "{YOUR_GITHUB_EMAIL}"
repo_url = f"https://{YOUR_GITHUB_USERNAME}:{GITHUB_PAT}@github.com/{GITHUB_OWNER_USERNAME}/{GITHUB_REPO_NAME}.git"
!rm -rf {GITHUB_REPO_NAME}
print(f"Cloning repository {GITHUB_OWNER_USERNAME}/{GITHUB_REPO_NAME}...")
!git clone {repo_url}

# Di chuyển vào trong thư mục repo
%cd {GITHUB_REPO_NAME}

# 2. CHẠY SCRIPT ĐỔI TÊN HÀNG LOẠT VỀ DẠNG 001.jpg
try:
    if os.path.isdir(TARGET_IMAGE_DIR):
        print(f"\nBắt đầu đổi tên các file trong: {TARGET_IMAGE_DIR}")
        # Lấy danh sách file và sắp xếp để đảm bảo thứ tự
        filenames = sorted(os.listdir(TARGET_IMAGE_DIR))
        counter = 1
        for filename in filenames:
            if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
                extension = os.path.splitext(filename)[1]
                # Format tên mới: 001, 002,...
                new_name = f"{counter:03d}{extension}"

                old_file_path = os.path.join(TARGET_IMAGE_DIR, filename)
                new_file_path = os.path.join(TARGET_IMAGE_DIR, new_name)

                os.rename(old_file_path, new_file_path)
                print(f"Đã đổi tên: {filename} -> {new_name}")
                counter += 1
        print(f"\nHoàn tất! Đã đổi tên {counter - 1} file.")
    else:
        print(f"[LỖI] Không tìm thấy thư mục '{TARGET_IMAGE_DIR}'.")
except Exception as e:
    print(f"Đã xảy ra lỗi: {e}")

# Quay trở lại thư mục gốc
%cd ..

# 3. TẠO COMMIT MỚI VÀ PUSH (Cách an toàn)
print("\n--- Bắt đầu tạo commit MỚI trên Git ---")
!cd {GITHUB_REPO_NAME} && \
  git add . && \
  git commit -m "Chore: Standardize image filenames to 001, 002 format" && \
  git push

print("--- Hoàn tất! Đã đẩy lên commit sửa tên file. ---")
```
